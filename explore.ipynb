```python
from threading import Timer


def debounce(wait):
    """ Postpone the execution of a function
        until after wait seconds elapsed since
        last time invoked. """
    def decorator(fn):
        def debounced(*args, **kwargs):
            def call_it():
                fn(*args, **kwargs)
            try:
                debounced.t.cancel()
            except(AttributeError):
                pass
            debounced.t = Timer(wait, call_it)
            debounced.t.start()
        return debounced
    return decorator
```
